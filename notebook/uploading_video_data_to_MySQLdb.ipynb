## adding data to sql database

```python
#function to read MySQL query output
def print_mysql_output(cursor):
    print(" -- START --")
    rows = cursor.fetchall()
    for r in rows:
        print(r)
    print(" -- END --\n")
```

```python
host = ""
user = ""
password = ""
```

```python
#creating connection to the MySQL database
db = conn.connect(
    host = host,
    user = user,
    passwd = password,
        )
```

```python
#creating cursor object
cursor = db.cursor()
```

```python
cursor.execute("show databases;")
print_mysql_output(cursor)
```

```python
cursor.execute("create database youtube_db;")
```

```python
cursor.execute("show databases;")
print_mysql_output(cursor)
```

```python
cursor.execute("use youtube_db;")
```

```python
#create table
cursor.execute("CREATE TABLE IF NOT EXISTS mytable (channel_title varchar(50), video_id varchar(50), Title varchar(255), Thumbnail_url varchar (255), Published_date varchar(50), Views int, Likes int, Comment_count int, video_url varchar(255));")
```

```python
#insert the dataframe data into table
for row in df.itertuples(index = False):
    cursor.execute("INSERT INTO mytable (channel_title, video_id, Title, Thumbnail_url, Published_date , Views , Likes , Comment_count, video_url) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)", (row.channel_Title, row.video_id, row.Title, row.Thumbnail_url, row.Published_date, row.Views, row.Likes, row.Comment_count, row.video_url))
#commit the changes to the database
db.commit()    
```

```python
cursor.execute("select * from mytable;")
print_mysql_output(cursor)
```

```python
#close the cursor and connection
cursor.close()
db.close()
```
