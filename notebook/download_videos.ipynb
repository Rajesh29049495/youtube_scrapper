```python
from pytube import YouTube, exceptions
import json
import os
import pandas as pd
```

```python
df = pd.read_csv("Video_details.csv")
```

```python
df
```

```python
# define the list of preferred video resolutions
preferred_resolutions = ['360p', '480p', '720p', '144p']

# define the directory to save the downloaded video
main_dir = "videos"

# define a list to keep track of skipped videos
skipped_videos= []

# looping through each video Id
for i, video_id in enumerate(df['video_id']):
    
    # define the name of the video file
    filename = f"{video_id}.mp4"
    
    # check if the video has already ben downloaded
    if os.path.isfile(os.path.join(main_dir, filename)):
        print(f"{i+1}: {filename} already exists.")
        continue
    
    # url of the video
    url = df['video_url'][i]             #url = f"https://www.youtube.com/watch?v={video_id}"
    
    # try to download the video at each preferred resolution
    for resolution in preferred_resolutions:
        try:
            #connect to the youtube video an ddownload it at the selected resolution
            print(f"{i+1}: connecting to {video_id}")
            yt_video = YouTube(url)
            stream = yt_video.streams.filter(progressive = True, file_extension = 'mp4', res = resolution).first()    
            stream.download(main_dir, filename = filename)
            print(f"{i+1}: downloaded {video_id} in {stream.resolution}")
            #exit the loop once the video has been downloaded
            break
            
        except (exceptions.LiveStreamError, StopIteration):
            print(f"{i+1}: skipped live stream {video_id}")
            skipped_videos.append(video_id)
            break
            
        except Exception as e:
            # handle other exceptions by trying the next preferred resolution
            print(f"{i+1}: error downloading {video_id} in {resolution}: {e}")
            continue
    else:
        # if all resolution fail, skip the video an dmove on to the next one
        print(f"{i+1}: failed to download {video_id}")
        skipped_videos.append(video_id)
    
```
